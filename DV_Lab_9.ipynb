# **Exploratory Data Analysis**

Exploratory Data Analysis (EDA) is a crucial step in the data analysis process. It involves the examination, visualization, and summary of key characteristics and patterns within a dataset to gain insights and inform further data analysis.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the dataset
iris_data = sns.load_dataset("iris")

# Step 2: Explore basic dataset information
num_rows, num_cols = iris_data.shape
print(f"Number of Rows: {num_rows}")
print(f"Number of Columns: {num_cols}")
print(iris_data.info())
print(iris_data.describe())

# Step 3: Handle missing values and duplicates (if any)
# For this dataset, we assume there are no missing values or duplicates.

# Step 4: Data visualization
plt.figure(figsize=(8, 6))
sns.histplot(data=iris_data, x="sepal_length", hue="species", kde=True)
plt.title("Distribution of Sepal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.show()

# Step 5: Investigate relationships between variables
sns.set_style("whitegrid")
sns.pairplot(iris_data, hue="species", height=2)
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()

# Step 6: Create summary statistics and insights
species_avg_sepal_length = iris_data.groupby("species")["sepal_length"].mean()
print("Average Sepal Length by Species:")
print(species_avg_sepal_length)
```
