### **Task 3: Given a json data, create a pandas dataframe from it**

```python
import pandas as pd
import json

# Sample JSON data for 5 countries
world_population_data = [
    {
        "country": "China",
        "population_2021": 1440000000,
        "population_2020": 1430000000,
        "population_growth_rate": 0.7
    },
    {
        "country": "India",
        "population_2021": 1390000000,
        "population_2020": 1380000000,
        "population_growth_rate": 1.2
    },
    {
        "country": "United States",
        "population_2021": 332000000,
        "population_2020": 331000000,
        "population_growth_rate": 0.5
    },
    {
        "country": "Brazil",
        "population_2021": 213000000,
        "population_2020": 212000000,
        "population_growth_rate": 0.9
    },
    {
        "country": "Russia",
        "population_2021": 145000000,
        "population_2020": 144000000,
        "population_growth_rate": 0.3
    }
]

# Create a Pandas DataFrame from the JSON data
df = pd.DataFrame(world_population_data)

# Display the DataFrame
print(df)
```

```python
# Visualization 1: Population of Each Country in 2021
plt.figure(figsize=(10, 6))
plt.bar(df['country'], df['population_2021'])
plt.xlabel('Country')
plt.ylabel('Population in 2021')
plt.title('Population of Each Country in 2021')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 2: Population Growth Rate
plt.figure(figsize=(8, 6))
plt.bar(df['country'], df['population_growth_rate'])
plt.xlabel('Country')
plt.ylabel('Population Growth Rate (%)')
plt.title('Population Growth Rate by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 3: Population Comparison (2021)
plt.figure(figsize=(10, 6))
plt.bar(df['country'], df['population_2021'], label='2021', alpha=0.7)
plt.bar(df['country'], df['population_2020'], label='2020', alpha=0.7)
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population Comparison (2020 vs. 2021)')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
