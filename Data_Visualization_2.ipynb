## **Visualization of Immigration to Canada data set**

## **Loading the dataset**

```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/Canada.csv')
df.info()
```

```python
df.head(5)
```

```python
df.info()
```

### **Data cleaning**

```python
# Task 1: Drop the unnecessary columns
df.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df.info()
```

```python
# Task 2: Rename the column names as 'OdName' -> 'Country', 'AreaName' -> 'Continent', 'RegName' -> 'Region'
df.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df.head()
```

```python
# Task 3: Drop the country 'Unknown'
df = df.drop(df.loc[df["Country"] == 'Unknown'].index)
df.set_index('Country', inplace=True)
```

```python
# Task 4: Drop the rows with nan indices

nan_index_mask = df.index.isna()

# Drop rows with NaN indices
df = df[~nan_index_mask]

# Drop last columns of nan
df.drop(["Unnamed: 43", "Unnamed: 44", "Unnamed: 45", "Unnamed: 46", "Unnamed: 47", "Unnamed: 48",
         "Unnamed: 49", "Unnamed: 50"], axis = 1, inplace = True)

df.head()
```

### **Data Visualization**

```python
# 1. How was the immigration from India to Canada from 1990 to 2013?
import matplotlib.pyplot as plt

years = list(map(str, range(1990, 2014)))

india = df.loc['India', years]

# Change the index of India to type integer for plotting
india.index = india.index.map(int)

ax = india.plot(kind='bar', figsize=(20,10))
ax.set(title='Immigration From India', ylabel='Number of Immigrants', xlabel='Years')
plt.show()
```

```python
# 2. Create an area plot for top 5 immigrant countries from 1990 to 2013

df['Total'] = df.sum(axis=1)
df.sort_values(by='Total', ascending = False, axis = 0, inplace = True)
df_top5 = df.head(5)
```

```python
df_top5
```

```python
df_top5 = df_top5[years].transpose()
df_top5.index = df_top5.index.map(int)
ax = df_top5.plot(kind = 'area', figsize=(20,10))
ax.set(title='Immigration Trend of Top 5 Countries',
       ylabel='Number of Immigrants',
       xlabel='Years');
```

```python
# 3. What is the number of new immigrants from the various countries to Canada in 2013?

bin_counts, bin_edges = np.histogram(df['2013'])

ax = df['2013'].plot(kind='hist', figsize=(20,10), xticks=bin_edges)
ax.set(title='Immigration From 195 Countries in 2013',
       ylabel='Number of Countries',
       xlabel='Number of Immigrants');
```

```python
# 4. What is the percentage of new immigrants grouped by continents from 1980 to 2013?
df_continents = df.groupby('Continent', axis=0).sum()

fig = plt.figure()
# Change background color
fig.patch.set_facecolor('white')

df_continents['Total'].plot(kind='pie',
                            figsize=(20, 10))

plt.title('Immigration to Canada by Continent From 1980 to 2013')
# Set equal scaling (i.e., make circles circular) by changing axis limits
plt.axis('equal')
plt.legend(labels=df_continents.index, loc='upper left');
```

```python
# 5. Compare the distribution of the number of new immigrants from India, Pakistan, China between 1980 and 2013.

df_ci = df.loc[['China','India', 'Pakistan'], years].transpose()

ax = df_ci.plot(kind='box', figsize=(20, 10))

ax.set(title='Immigrants From China and India From 1980 to 2013',
       xlabel="Country",
       ylabel='Number of Immigrants');
```
