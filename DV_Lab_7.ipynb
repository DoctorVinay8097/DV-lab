# **Heatmaps**

A **heatmap** is a data visualization technique that uses a color-coded matrix to represent the values of a two-dimensional dataset. It is particularly useful for visualizing relationships, patterns, and variations in data across rows and columns. Heatmaps are widely used in various fields, including statistics, data analysis, and data science, to provide insights into data by emphasizing differences and similarities between data points.

You are given a dataset containing information about the relationships between two variables over time. Your task is to create a heatmap to visualize the correlation between these variables and understand how they change across different time periods.

You will use the built-in "flights" dataset from Seaborn, which contains the number of passengers flying on various airlines over different months.

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the dataset
# We'll use the built-in "flights" dataset from Seaborn
flights_data = sns.load_dataset("flights")

# Step 2: Explore the dataset
# Check the first few rows to understand its structure
print(flights_data.head())

# Step 3: Create a heatmap to visualize the correlation
# We'll pivot the data to create a matrix suitable for a heatmap
correlation_matrix = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt='d', cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Passengers by Year and Month")
plt.show()

# Step 4: Analyze and interpret the heatmap
# The heatmap shows the correlation between the number of passengers in different months across the years.
# Brighter cells indicate a higher number of passengers, and darker cells indicate lower numbers.
# It's clear that there is a positive correlation and a seasonal pattern, with more passengers during certain months.

# Questions and Answers:

# Q1: How can you customize the color map of the heatmap to use a different colormap (e.g., "viridis")?
# A1: You can customize the colormap by setting the cmap parameter when creating the heatmap. Here's the code to use the "viridis" colormap:
sns.heatmap(correlation_matrix, annot=True, fmt='d', cmap='viridis', linewidths=0.5)

# Q2: How can you add x-axis and y-axis labels to the heatmap to label the months and years?
# A2: You can add x-axis and y-axis labels using the xticklabels and yticklabels parameters. Here's the code to add labels:
sns.heatmap(correlation_matrix, annot=True, fmt='d', cmap='coolwarm', linewidths=0.5,
            xticklabels=flights_data['year'].unique(), yticklabels=flights_data['month'].unique())

# Q3: How can you set a specific figure size for the heatmap to make it larger?
# A3: You can set the figure size using the plt.figure(figsize=(width, height)) function. Here's the code to set a larger figure size:
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='d', cmap='coolwarm', linewidths=0.5)

# Q4: How can you adjust the font size of the annotations in the heatmap?
# A4: You can adjust the font size of the annotations using the annot_kws parameter. Here's the code to increase the font size:
sns.heatmap(correlation_matrix, annot=True, fmt='d', cmap='coolwarm', linewidths=0.5,
            annot_kws={"size": 12})

# Q5: How can you save the heatmap as an image file (e.g., PNG) in your local directory?
# A5: You can save the heatmap as an image using the plt.savefig() function. Here's the code to save it as a PNG file:
sns.heatmap(correlation_matrix, annot=True, fmt='d', cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Passengers by Year and Month")
plt.savefig("heatmap.png")
```
